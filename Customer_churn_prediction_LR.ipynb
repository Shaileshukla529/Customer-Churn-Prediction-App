# CREATE PREPROCESSING PIPELINE

```python
y =df['Churn'].copy()
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
X = df.drop('Churn',axis = 1)
X.head()
```

### Splitting data set into training set and test set 

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### Define preprocessing for numerical columns

```python
numerical_transformer = StandardScaler()
```

### Define preprocessing for Binary columns

```python
""" For binary columns (like Yes/No features) with OneHotEncoder(drop='first'), you'll end up with a single column with 0/1 values - which is effectively the same as what LabelEncoder would produce.
For example, if 'Partner' has values 'Yes' and 'No', after transformation:

With OneHotEncoder(drop='first'), you get one column 'Partner_Yes' (0 if 'No', 1 if 'Yes')
With LabelEncoder(), you might get 'Partner' (0 if 'No', 1 if 'Yes') """

binary_transformer = OneHotEncoder(drop ="first") 
```

### Define preprocessing for nominal categorical columns

```python
nominal_transformer = OneHotEncoder(drop ="first")
```

### Define preprocessing for ordinal categorical columns 

```python
contract_categories = [['Month-to-month', 'One year', 'Two year']]
ordinal_transformer = OrdinalEncoder(categories=contract_categories)
```

## Creating preprocessing steps

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_cols),
        ('int', 'passthrough', integer_cols),  # Already 0/1
        ('bin', binary_transformer, binary_cols),
        ('nom', nominal_transformer, nominal_cols),
        ('ord', ordinal_transformer, ordinal_cols)
    ]
)
```

## Create pipeline with preprocessing and model

```python
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42))
])
```

#### Training the model

```python
pipeline.fit(X_train, y_train)
```

#### Making predictions

```python
y_pred = pipeline.predict(X_test)
```

#### Evaluating the model

```python
print("\nClassification Report:  Logistic Regrssion")
print(classification_report(y_test, y_pred))
```

```python
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
print("\nConfusion Matrix:  Logistic Regrssion")
conf_matrix = confusion_matrix(y_test, y_pred)
conf_matrix
```

### Hyperparameter Tunning

```python
# Define parameter grids with VALID combinations for Logistic Regression

# Grid 1: For 'liblinear' solver (supports L1 and L2)
param_grid_liblinear = {
    'classifier__penalty': ['l1', 'l2'],
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'classifier__solver': ['liblinear'],
    'classifier__max_iter': [100, 200] # Example iter values
}

# Grid 2: For 'lbfgs' solver (supports L2 and None)
param_grid_lbfgs = {
    'classifier__penalty': ['l2', None], # Use None, not 'none'
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'classifier__solver': ['lbfgs'],
    'classifier__max_iter': [100, 200, 300] # Example iter values
}

# Grid 3: For 'saga' solver (supports L1, L2, ElasticNet, None)
param_grid_saga = {
    'classifier__penalty': ['l1', 'l2', 'elasticnet', None], # Use None
    'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
    'classifier__solver': ['saga'],
    'classifier__l1_ratio': [0, 0.25, 0.5, 0.75, 1], # Only used when penalty='elasticnet'
    'classifier__max_iter': [100, 200, 300, 400] # Saga might need more iterations
}

# Combine the grids into a list
combined_param_grids = [
    param_grid_liblinear,
    param_grid_lbfgs,
    param_grid_saga
]

# Note: The 'valid_samples' list generation from your original cell [80] is not needed
# when using RandomizedSearchCV with param_distributions.
```

```python
random_search = RandomizedSearchCV(
    estimator = pipeline,
    param_distributions= combined_param_grids, # **** USE THE LIST OF VALID GRIDS ****
    n_iter= 50,  # Number of parameter settings that are sampled. n_iter trades off runtime vs quality of the solution.
    cv=5,
    scoring= 'accuracy',
    n_jobs=-1,
    verbose = 1,
    error_score=0, # Keep as 0 or 'warn', though errors should be less frequent now
    random_state=42
)
```

```python
random_search.fit(X_train, y_train)
```

```python
print("Best Parameters:", random_search.best_params_)
print("Best CV Score:", random_search.best_score_)
```

```python
# Evaluate on test set
best_model = random_search.best_estimator_
y_pred = best_model.predict(X_test)
print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```
